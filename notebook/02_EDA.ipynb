```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
churn = pd.read_csv('churn.csv')
churn.head()

# Histogram
def plot_histogram(churn , col_name) :
  plt.figure(figsize =(5 , 3))
  sns.histplot(churn[col_name] , kde = True)
  plt.title(f"{col_name} Distribution")
  plt.ylabel("Count")
  plt.xlabel(col_name)
  col_mean = churn[col_name].mean()
  col_median = churn[col_name].median()

  plt.axvline(col_mean , color = "red" , linestyle = "dashed" , linewidth = 2 , label = "mean")
  plt.axvline(col_median , color = "green" , linestyle = "dashed" , linewidth = 2 , label = "median")
  plt.legend()

  plt.show()
```

```python
plot_histogram(churn , "tenure")
plot_histogram(churn , "MonthlyCharges")
```

```python
#BoxPlot
def plot_boxplot(df , col_name) :
  plt.figure(figsize =(5 , 3))
  sns.boxplot(y=churn[col_name])
  plt.title(f"{col_name} boxplot")
  plt.show()
```

```python
plot_boxplot(churn , 'MonthlyCharges')
plot_boxplot(churn , 'tenure')
```

```python
#heatmap/Corelation
numeric_df = churn.select_dtypes(include=['float64', 'int64'])
sns.heatmap(numeric_df.corr(), annot=True)
plt.show()
```
