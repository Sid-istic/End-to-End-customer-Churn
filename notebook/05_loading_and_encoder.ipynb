```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
import pickle
from sklearn.ensemble import RandomForestClassifier
```

```python
sample_raw = {
    'gender': 'Male',             # Gender typically has less impact on churn
    'SeniorCitizen': 1,           # Senior citizens are sometimes slightly more prone to churn
    'Partner': 'No',              # Not having a partner might indicate a less stable situation
    'Dependents': 'No',           # Not having dependents might also indicate a less stable situation
    'tenure': 5,                  # Short tenure (e.g., 5 months)
    'PhoneService': 'Yes',
    'MultipleLines': 'No',        # Single phone line
    'InternetService': 'Fiber optic', # Fiber optic users sometimes have higher churn rates than DSL
    'OnlineSecurity': 'No',       # Does NOT have online security
    'OnlineBackup': 'No',         # Does NOT have online backup
    'DeviceProtection': 'No',     # Does NOT have device protection
    'TechSupport': 'No',          # Does NOT have tech support - often a strong indicator of churn risk
    'StreamingTV': 'Yes',         # Still uses some services
    'StreamingMovies': 'Yes',     # Still uses some services
    'Contract': 'Month-to-month', # Month-to-month contract - strong indicator of churn risk
    'PaperlessBilling': 'Yes',    # Paperless billing users are sometimes more likely to churn
    'PaymentMethod': 'Electronic check', # Electronic check - strong indicator of churn risk
    'MonthlyCharges': 90.5,       # Relatively high monthly charge for limited support services
    'TotalCharges': 452.5         # Total charges consistent with short tenure and monthly charges (5 * 90.5)
}

input_data_df = pd.DataFrame([sample_raw])
input_data_df
```

```python
with open("encoder.pkl", "rb") as f:
    encoder = pickle.load(f)
```

```python
for col,encode in encoder.items():
    input_data_df[col] = encode.transform(input_data_df[col])
```

```python
input_data_df
```

```python
with open("Random Forest.pkl", 'rb') as f:
    loaded_model = pickle.load(f)

```

```python
prediction = loaded_model.predict(input_data_df)
```
